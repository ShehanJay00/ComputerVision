```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../DATA/00-puppy.jpg')
# We can read a Image directly as a Numpy Array using opencv
```

```python
img
```

```python
img_bgr = cv2.imread('../DATA/00-puppy.jpg')
plt.imshow(img_bgr)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

```python
img_gray.shape
```

## Resize the Image

```python
img_rgb.shape
```

```python
img =cv2.resize(img_rgb,(1000,500))
plt.imshow(img)
```

## Resize the Image by Ratio

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(img_rgb, (0, 0), fx=w_ratio, fy=h_ratio)
plt.imshow(new_img)
```

## Flip the Image

```python
new_img = cv2.flip(new_img,0)
# Flip Horizontally
plt.imshow(new_img)
```

```python
new_img = cv2.flip(new_img,1)
# Flip Vertically
plt.imshow(new_img)
```

```python
new_img = cv2.flip(new_img,-1)
plt.imshow(new_img)
```

## Save Image Files

```python
cv2.imwrite('my_new_picture.jpg',new_img)
```

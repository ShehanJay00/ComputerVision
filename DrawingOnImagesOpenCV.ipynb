## Drawing on Images

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Create an blank Image
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

## Draw Shapes

```python
# pt1 = top left corner
# pt2 = bottom right corner
cv2.rectangle(blank_img,pt1=(384,10),pt2=(500,100),color=(0,255,0),thickness=5)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=5)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(400,400), radius=50, color=(255,0,0), thickness=5)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(100,100), radius=50, color=(255,0,0), thickness=-1)
plt.imshow(blank_img)
```

```python
# Draw a diagonal blue line with thickness of 5 px
cv2.line(blank_img,pt1=(0,0),pt2=(511,511),color=(102, 255, 255),thickness=5)
plt.imshow(blank_img)
```

## Text

```python
font = cv2.FONT_HERSHEY_COMPLEX
```

```python
cv2.putText(blank_img,text='Hello',org=(10,500), fontFace=font,fontScale= 4,color=(255,255,255),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```
